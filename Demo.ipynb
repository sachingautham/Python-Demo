### **Palindrome function**

```python
 def palindrome(S):

   i=0
   while i<=len(S)/2:
     if S[i] != S[-i-1]:

       return False
     i+=1
   return True 
```

```python
palindrome("RACECAR")
```

```python
Palindromes = [ ]

for a in range(10,100):
  for b in range(10,100):
    strnum = str(a*b)
    if palindrome(strnum) == True:
      Palindromes.append(a*b)
max(Palindromes)
```

```python
x_val = np.linspace(1,20,80)
y_val = np.linspace(1,20,80)
X,Y = np.meshgrid(x_val,y_val)
def Z(x,y):
  return np.sin(x**2 + y**2) / (x**2 + y**2)
plt.contour(X, Y,Z(X,Y), colors='black')
plt.contour(X,Y, Z(X,Y), cmap='RdGy')
plt.colorbar()
```
